```python
# Import the libraries
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
import cv2

# Read the image file
img=cv2.imread(r"C://Users//vinay//Desktop//bharat gpt-venvv//bharat//image2.jpg")

# Convert the image to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply some preprocessing to improve the OCR accuracy
# You can experiment with different methods and parameters
img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Recognize the text from the image using PyTesseract
text = pytesseract.image_to_string(img)

# Print the text to the console
print(text)

# Save the text to a text file
with open("output.txt", "w") as f:
    f.write(text)
```

```python
# Import the libraries
import pytesseract
from pdf2image import convert_from_path

# Specify the path of the PDF file
PDF_file = "C://Users//vinay//Desktop//bharat gpt-venvv//bharat//sample.pdf"

# Convert the PDF file into a list of images
images = convert_from_path(PDF_file,500,poppler_path="C:\Program Files\poppler-23.11.0\Library\bin")

for i in range(len(images)):
    images[i].save('pdfimg'+str(i)+'.jpg','JPEG')
# Initialize an empty string to sto/re the text
print("successfully created the images")
text = ""

# Loop through the images and append the text to the string
for image in images:
    # Recognize the text from the image using PyTesseract
    text += pytesseract.image_to_string(image)
    print(text)
    print("======================")
# Save the text into a text file
with open("output.txt", "w") as f:
    f.write(text)
```
